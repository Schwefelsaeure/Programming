## Imports

```python
import pandas as pd
import numpy as np
from  matplotlib import pyplot as plt
```

## Read Data

```python
# data = pd.read_csv("potentials/target1_potential.csv")
data = pd.read_csv("potentials/bruteforce_potential.csv")

x_samples = len(data.loc[data["y"] == data["y"].min()])
y_samples = len(data.loc[data["x"] == data["x"].min()])

print("x_samples: {}".format(x_samples))
print("y_samples: {}".format(y_samples))
```

## Reshape Data to Fit for pyplot.contour()

```python
x_reshaped = data["x"].values.reshape(-1, x_samples)
y_reshaped = data["y"].values.reshape(-1, x_samples)
values_reshaped = data["value"].values.reshape(-1, x_samples)

plt.contour(x_reshaped, y_reshaped, values_reshaped)
```
