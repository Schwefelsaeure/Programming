```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

total_samples = 200
np.random.seed(444)

data = np.random.randint(0, 100, size=(2, total_samples))

fig, axes = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = axes

ax1.hist(data[0])
ax1.set_title("Histogram: $data[0]$")

ax2.scatter(x=data[0], y=data[1], color="green")
ax2.set_title("Scatter: $data[0]\ vs. data[1]$")
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 100)
ax2.yaxis.tick_right()

# fig.savefig("MyFigure.pdf")
```
