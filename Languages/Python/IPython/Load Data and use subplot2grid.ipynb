```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

import tarfile

gridsize = (3, 2)
fig = plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2)
ax2 = plt.subplot2grid(gridsize, (1, 0), colspan=2)
ax3 = plt.subplot2grid(gridsize, (2, 0))
ax4 = plt.subplot2grid(gridsize, (2, 1))

fpath = 'CaliforniaHousing/cal_housing.data'
with tarfile.open("./Downloads/cal_housing.tgz", mode='r') as archive:
    data = np.loadtxt(archive.extractfile(fpath), delimiter=',')
    
y = data[:, -1]
pop, age = data[:, [4, 7]].T

ax1.set_title('Home value as a function of home age & area population', fontsize=12)
sctr = ax1.scatter(x=age, y=pop, c=y, cmap='RdYlGn')
plt.colorbar(sctr, ax=ax1, format='$%d')
ax1.set_yscale('log')
ax2.scatter(x=age, y=y)
ax3.hist(age, bins='auto')
ax4.hist(pop, bins='auto', log=True)
fig.show()
```
